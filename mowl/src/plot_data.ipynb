```python
year = 2018
cn_nst = pd.read_excel(f'../data/correspondence_tables/NST_2007_CN_{year}.xlsx', dtype={'NST 2007': str, f'CN {year}': str}, usecols=['NST 2007', f'CN {year}'])
cn_nst
```

```python
cn_nst.rename(columns={'NST 2007': 'NST', f'CN {year}': 'CN'}, inplace=True)
cn_nst
```

Dataframe linking NST2007 to CE25

```python
res2 = list(default_world.sparql("""
                PREFIX owl: <http://www.w3.org/2002/07/owl#>
                PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
                PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
                PREFIX nst: <http://www.semanticweb.org/geofluxus/ontologies/2022/NST2007#>        
                PREFIX cn: <http://www.semanticweb.org/geofluxus/ontologies/2022/CN2021#>        
                PREFIX ce: <http://www.semanticweb.org/geofluxus/ontologies/2022/CE25#>
                PREFIX mat: <http://www.semanticweb.org/geofluxus/ontologies/2022/dummy_materials#>
        
                SELECT DISTINCT ?c
                WHERE {
                  ?c rdfs:subClassOf ce:CE25 .
                }
                """))

data = {'CE': [], 'NST': []}
for r in res2:
  # print(r[0].CE25Code[0], r[0].NSTCode)
  for nst in sorted(r[0].NSTCode):
    data['CE'].append(r[0].CE25Code[0])
    data['NST'].append(nst)

df_full = pd.DataFrame.from_dict(data)
df_full
```

```python
df_final = df_full.merge(cn_nst, on='NST', how='left', indicator=True)
df_final
```

<font style="color: green; font-weight: bold">PROCESSING </font>Material's Data

```python
df_mat_data = pd.read_excel('../data/output_v2.xlsx', dtype={'NST 2007': str, 'CN_Code': str}, usecols=['NST 2007', 'CN_Code', 'hasData'])
df_mat_data
```

```python
df_raw = df_final[['CE','NST','CN']]
df_mat_data.rename(columns={'CN_Code':'CN', 'NST 2007':'NST'}, inplace=True)
```

```python
df_raw['CN6'] = df_raw['CN'].str[:7]
df_raw
```

Format CN codes in Material's data

```python
df_mat_data['CN'] = df_mat_data['CN'].str[:4]+' '+df_mat_data['CN'].str[4:6]+' '+df_mat_data['CN'].str[6:8]
df_mat_data
```

<font style="color: green; font-weight: bold">PROCESSING </font>Merging Material's data with NST/CE

```python
df_raw = df_raw.merge(df_mat_data[['CN','hasData']], on='CN', how='left', indicator=True)
df_raw
```

```python
# assign colours to dataframe
df_raw['colors'] = df_raw['hasData'].map({True:'goldenrod', False:'lightgray'})
df_raw
```

<font style="color: cyan; font-weight: bold">IMAGE -- </font>New Plots

```python
def plotCE(code, height=3000, dataframe=df_raw):
  temp = dataframe[dataframe['CE'] == code]
  temp_fig = px.parallel_categories(temp,  
                                labels={'CN':'CN Codes', 'CN6':'CN Codes (6 digits)', 'NST':'NST Codes', 'CE': 'CE Codes'},
                                dimensions=['CN', 'CN6', 'NST', 'CE'],
                                color='colors',
                                title=f'All CN {year} and NST codes associated with CE25 code "{code}"',
                                height=height
                              )
  temp_fig.show()
  return temp_fig
```

```python
img = plotCE('25', height=3200)
```
