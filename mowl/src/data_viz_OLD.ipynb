```python
res1 = list(default_world.sparql("""
                PREFIX owl: <http://www.w3.org/2002/07/owl#>
                PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
                PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
                PREFIX nst: <http://www.semanticweb.org/geofluxus/ontologies/2022/NST2007#>        
                PREFIX cn: <http://www.semanticweb.org/geofluxus/ontologies/2022/CN2021#>        
                PREFIX ce: <http://www.semanticweb.org/geofluxus/ontologies/2022/CE25#>
                PREFIX mat: <http://www.semanticweb.org/geofluxus/ontologies/2022/dummy_materials#>
        
                SELECT DISTINCT ?cn ?nst ?ce
                WHERE {
                  ?s1 rdfs:subClassOf mat:Material .
                  ?s2 a ?s1;
                    mat:CNCode ?cn ;
                    mat:NSTCode ?nst ;
                    mat:CECode ?ce .
                }
                ORDER BY ASC(?cn)
                """))

print(res1)
```

Load Ontologies from quadstore

```python
onto_cn  = default_world.get_ontology('http://www.semanticweb.org/geofluxus/ontologies/2022/CN2021').load()
onto_nst = default_world.get_ontology('http://www.semanticweb.org/geofluxus/ontologies/2022/NST2007').load()
onto_ce = default_world.get_ontology('http://www.semanticweb.org/geofluxus/ontologies/2022/CE25').load()
```

```python
cn_nst = pd.read_excel('../data/correspondence_tables/NST_2007_CN_2021.xlsx', dtype={'NST 2007': str, 'CN 2021': str}, usecols=['NST 2007', 'CN 2021'])
cn_nst
```

```python
nsts = cn_nst['NST 2007'].unique()
print(len(nsts), nsts)
```

```python
dict_nst_cn = {}
for n in nsts:
  slc =  cn_nst[cn_nst['NST 2007'] == n]
  dict_nst_cn[n] = [cn for cn in  slc['CN 2021'].unique()]
```

```python
print(dict_nst_cn)
```

```python
cn_nst.rename(columns={'NST 2007': 'NST', 'CN 2021': 'CN'}, inplace=True)
cn_nst
```

```python
res2 = list(default_world.sparql("""
                PREFIX owl: <http://www.w3.org/2002/07/owl#>
                PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
                PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
                PREFIX nst: <http://www.semanticweb.org/geofluxus/ontologies/2022/NST2007#>        
                PREFIX cn: <http://www.semanticweb.org/geofluxus/ontologies/2022/CN2021#>        
                PREFIX ce: <http://www.semanticweb.org/geofluxus/ontologies/2022/CE25#>
                PREFIX mat: <http://www.semanticweb.org/geofluxus/ontologies/2022/dummy_materials#>
        
                SELECT DISTINCT ?c
                WHERE {
                  ?c rdfs:subClassOf ce:CE25 .
                }
                """))

data = {'CE': [], 'NST': []}
for r in res2:
  # print(r[0].CE25Code[0], r[0].NSTCode)
  for nst in sorted(r[0].NSTCode):
    data['CE'].append(r[0].CE25Code[0])
    data['NST'].append(nst)

df_full = pd.DataFrame.from_dict(data)
df_full
```

```python
df_final = df_full.merge(cn_nst, on='NST', how='left', indicator=True)
df_final
```

```python
uniq_colours = df_final['NST'].unique()

import random
rand = random.sample(range(len(total_colors)), len(uniq_colours))
selection = [total_colors[i] for i in rand]
print(len(selection), selection)
```

```python
color_dict_full = {}
for i in range(len(uniq_colours)):
  color_dict_full[uniq_colours[i]] = selection[i]

print(color_dict_full)
```

```python
df_final['colors'] = df_final['NST'].map(color_dict_full)
```

```python
df_f_01 = df_final[df_final['CE'] == '01']
fig_f01 = px.parallel_categories(df_f_01,  
                              labels={'CN':'CN Codes', 'NST':'NST Codes', 'CE': 'CE Codes'},
                              dimensions=['CN', 'NST', 'CE'],
                              color='colors',
                              title='All CN and NST codes associated with CE25 code "01"',
                              height=4000
                            )
fig_f01.show()
```

```python
df_raw = df_final[['CE','NST','CN']]
df_mat_data.rename(columns={'CN_Code':'CN', 'NST 2007':'NST'}, inplace=True)
```

```python
df_raw['CN6'] = df_raw['CN'].str[:7]
df_raw
```

```python
# print(df_mat_data['CN'].str[:4]+' '+df_mat_data['CN'].str[4:6]+' '+df_mat_data['CN'].str[6:8])
df_mat_data['CN'] = df_mat_data['CN'].str[:4]+' '+df_mat_data['CN'].str[4:6]+' '+df_mat_data['CN'].str[6:8]
df_mat_data
```

<font style="color: green; font-weight: bold">PROCESSING </font>Merging Material's data with NST/CE

```python
df_raw2 = df_raw.merge(df_mat_data[['CN','hasData']], on='CN', how='left', indicator=True)
df_raw2
```

```python
# assign colours to dataframe
df_raw2['colors'] = df_raw2['hasData'].map({True:'goldenrod', False:'lightgray'})
df_raw2
```

<font style="color: cyan; font-weight: bold">IMAGE -- </font>New Plots

```python
def plotCE(code, height=3000, dataframe=df_raw2):
  temp = dataframe[dataframe['CE'] == code]
  temp_fig = px.parallel_categories(temp,  
                                labels={'CN':'CN Codes', 'CN6':'CN Codes (6 digits)', 'NST':'NST Codes', 'CE': 'CE Codes'},
                                dimensions=['CN', 'CN6', 'NST', 'CE'],
                                color='colors',
                                title=f'All CN and NST codes associated with CE25 code "{code}"',
                                height=height
                              )
  temp_fig.show()
  return temp_fig
```
