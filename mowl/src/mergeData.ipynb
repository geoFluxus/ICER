# Merge NST2007 and GN2018 data

```python
import pandas as pd
import os

root_path = os.getcwd()
```

```python
src_path = os.path.dirname(root_path)
os.chdir(src_path)
data_path = os.getcwd()

path_cn2018 = data_path + r'\data\CNLijstVoorTNO_v5.xlsx'
path_NST2007 = data_path + r'\data\NST_2007_CN_2018 one-to-one correspondence 22-3-2018.xlsx'
sheet_cn2018 = 'CN2018_clean'

# check
print('file check: ', os.path.isfile(path_cn2018))
print('file check: ', os.path.isfile(path_NST2007))
```

```python

# data of interest in CN 2018 
df =  pd.read_excel(path_cn2018, sheet_name=sheet_cn2018)
# add leading zeros that are removed when python reads the content as int64
df['gn_code'] = df['gn_code'].astype('string').apply(lambda x: x.zfill(8))
col_cn2018 = df.columns

df
```

```python
# data for correspondence with NST 2007
df2 = pd.read_excel(path_NST2007)
df2['gn_code'] = df2['gn_code'].str.replace(' ','')
df2['gn_code'] = df2['gn_code'].astype('string').apply(lambda x: x.zfill(8))
col_nst2007 = df2.columns

df2
```

```python
df_merged = df.merge(df2, on='gn_code', how='left', indicator=True)

df_merged
```

```python
unmerged = df_merged[~df_merged['_merge'].str.contains('both')]

unmerged
```

```python
# check for missing data
for index, row in df2.iterrows():
  val = row['gn_code']
  if val in unmerged['gn_code'].values:
    print(val)
```

```python
writer = pd.ExcelWriter(data_path + r'\data\Analysis_by_NST2007.xlsx', engine='xlsxwriter')
suffixes = [str(i).zfill(2) for i in range(1,21)]
suffixes.append('NaN')

funcs = ['count_total', '#nan', '#zeros', 'min', 'max', 'avg_w_zeros', 'avg_no_zeros', 'total']
j=0
for d in lst_dfs:
  base = pd.DataFrame(funcs, columns=['Funcs'])
  temp = d.iloc[:, 3:66]
  cols = temp.columns

  for i in range(len(cols)):
    column = temp[cols[i]]
    count_nan = column.isnull().sum()
    count_zeros = (column == 0).sum()

    column = column.fillna(0)
    count = column.count()

    if (column == 0).sum() == column.size:
      min_val = 0
      avg_val_no0 = 0

    else:
      noZeros = column[column > 0]
      min_val = noZeros.min()
      avg_val_no0 = noZeros.mean()

    max_val = column.max()
    avg_val = column.mean()
    tot_val = column.sum()

    base[cols[i]] = [count, count_nan, count_zeros, min_val, max_val, avg_val, avg_val_no0, tot_val]
    base.to_excel(writer, sheet_name=f'nst_{suffixes[j]}', index=False)
  
  j+=1

writer.save()
```
