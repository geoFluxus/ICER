```python
import pandas as pd

path_folder = '../data/correspondence_tables/'
path_2018 = f'{path_folder}NST_2007_CN_2018.xlsx'
path_2019 = f'{path_folder}NST_2007_CN_2019.xlsx'
path_2020 = f'{path_folder}NST_2007_CN_2020.xlsx'
path_2021 = f'{path_folder}NST_2007_CN_2021.xlsx'

df_2018 = pd.read_excel(path_2018, dtype={'NST 2007': str, 'CN 2018': str}, usecols=['NST 2007', 'CN 2018'])
df_2019 = pd.read_excel(path_2019, dtype={'NST 2007': str, 'CN 2019': str}, usecols=['NST 2007', 'CN 2019'])
df_2020 = pd.read_excel(path_2020, dtype={'NST 2007': str, 'CN 2020': str}, usecols=['NST 2007', 'CN 2020'])
df_2021 = pd.read_excel(path_2021, dtype={'NST 2007': str, 'CN 2021': str}, usecols=['NST 2007', 'CN 2021'])
df_2018
```

```python
path_en = '../data/NST_2007_EN.csv'
df_raw = pd.read_csv(path_en, dtype={'Code': str, 'Parent': str}, usecols=['Code', 'Parent', 'Description'])
df_raw = df_raw.dropna().rename(columns={'Code': 'NST 2007'})

df_raw
```

```python
checker = {}
checker_2 = {}

for i in range(df_raw.shape[0]):
  code = df_raw.iloc[i, 0]
  cn_codes = df_2018[df_2018['NST 2007'].isin([code])]
  #print(code, '--', cn_codes['CN 2018'].tolist())
  cn_raw = cn_codes['CN 2018'].tolist()
  checker[code] = sorted(cn_raw)

  #cn_proc = list(set([c.split(' ')[0] for c in cn_raw]))
  cn_proc = list(set([c[:-3] for c in cn_raw]))
  checker_2[code] = sorted(cn_proc)
```

```python
from owlready2 import *
import types

# create and populate ontology
onto_ce25 = get_ontology('http://www.semanticweb.org/geofluxus/ontologies/2022/CE25')
print(onto_ce25.base_iri)
```

```python
# onto_nst = get_ontology('../ontologies/nst2007.owl').load()
# print(onto_nst.base_iri)

# onto_ce25.imported_ontologies.append(onto_nst)
```

```python
def assignNSTCode(name, code):
  name.NSTCode.append(code)

with onto_ce25:
  onto_ce25.metadata.versionInfo.append('1.1.0')
  onto_ce25.metadata.label.append('CE 25 Classification')
  onto_ce25.metadata.seeAlso.append('https://www.geofluxus.com/')
  class CE25(Thing):
    pass
    
  CE25.comment = [
    locstr(f'CE 25 Goederengroepen Circulaire Economie', lang='nl')
  ]
  CE25.label = [locstr('CE25 - Goederengroepen Circulaire Economie', lang='nl')]
  
  class NSTCode(DataProperty):
    range = [str]
  
  # class forNST(ObjectProperty):
  #   pass
  
  class CE25Code(DataProperty):
    range = [str]

  for index, row in df_ce25.iterrows():
    code_ce = str(row['CE25']).zfill(2)
    ce_desc = row['CE25_Description']
    nst_lst = row['NST_Groups']

    code = types.new_class(f'CE_{code_ce}', (CE25,))
    code.comment = [
      locstr(ce_desc, lang='nl'),
    ]
    code.label = [
      locstr(f'{code_ce} - {ce_desc}', lang='nl')
    ]
    code.CE25Code = [code_ce]

    for nst in nst_lst:
      # nst = 'NST_' + nst.replace('.','_')
      # print(onto_nst[nst].iri)
      assignNSTCode(code, nst)
      # code.forNST.append(onto_nst[nst])
```
